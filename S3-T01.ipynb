```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
dataFlights = pd.read_csv("DelayedFlights.csv")
```

```python
dataFlights.head(10)
```

##   Una variable categòrica (UniqueCarrier)

```python
plt.figure(figsize=(13,5))
sns.set_style("darkgrid")
sns.countplot(x = 'UniqueCarrier',data=dataFlights,palette='muted')
plt.title('Vols per cada aerolínea', size=21)
plt.ylabel('número de vols',size=13)
plt.xlabel('Codí d´aerolínea',size=13)
plt.savefig('Vols-Aerolineas.png',dpi=300,bbox_inches='tight')
```

##   Una variable numèrica (ArrDelay)

```python
plt.figure(figsize=(13,5))
sns.histplot(dataFlights['ArrDelay'])
plt.title('Retards d´arribada', size=21)
plt.ylabel('número de vols',size=13)
plt.xlabel('Retards d´arribada [min]',size=13)
```

#### Es limita l'eix x per poder analitzar els casos de retards típics. Menyspreem els casos de més de 400 minuts per no ser retards per causes normals. Els valors negatius corresponents a arribadas abans d'hora.

```python
plt.figure(figsize=(13,5))
sns.histplot(dataFlights['ArrDelay'])
plt.title('Retards d´arribada', size=21)
plt.xlim(-50,380)
plt.ylabel('número de vols',size=13)
plt.xlabel('Retards d´arribada [min]',size=13)
plt.savefig('Retards_arribada.png',dpi=300,bbox_inches='tight')
```

```python
plt.figure(figsize=(13,5))
sns.distplot(dataFlights['ArrDelay'],bins=500)
plt.xlim(-50,380)
plt.title('Retards d´arribada', size=21)
plt.savefig('Retards_arribada_distribucio.png',dpi=300,bbox_inches='tight')
```

#### Els mayor nombre de retards son de menys de 15 min

## Una variable numèrica i una categòrica (ArrDelay i UniqueCarrier)

```python
plt.figure(figsize=(21,13))
sns.boxplot(x='ArrDelay',y='UniqueCarrier',data=dataFlights)
plt.xlabel('Retards d´arribada [min]',size=16)
plt.ylabel('Aerolínea',size=16)
plt.savefig('Retards_arribada_vs_aerolineas_Boxplot.png',dpi=300,bbox_inches='tight')
```

```python
plt.figure(figsize=(21,13))
sns.violinplot(x='ArrDelay',y='UniqueCarrier',data=dataFlights)
plt.xlabel('Retards d´arribada [min]',size=16)
plt.ylabel('Aerolínea',size=16)
plt.savefig('Retards_arribada_vs_aerolineas_Violinplot.png',dpi=300,bbox_inches='tight')
```

## Dues variables numèriques (ArrDelay i DepDelay)

```python
joint_kws = dict(gridsize=100)
plot = sns.jointplot(x='ArrDelay',y='DepDelay',data=dataFlights,kind='hex',joint_kws=joint_kws)
plot.ax_marg_x.set_xlim(-50, 400)
plot.ax_marg_y.set_ylim(-50, 400)
plt.show()
fig=plot.fig
fig.savefig('Retards_arribada_vs_sortida_jointplot.png',dpi=300,bbox_inches='tight')
```

#### A simple vista es veu una relació lineal entre les dues variables.

```python
sns.lmplot(x='ArrDelay',y='DepDelay',data=dataFlights,aspect=1.5,height=10,scatter_kws={'s':5})
plt.savefig('Retards_arribada_vs_sortida.png',dpi=300,bbox_inches='tight')
```

## Tres variables (ArrDelay, DepDelay i UniqueCarrier)

```python
sns.lmplot(x='ArrDelay',y='DepDelay',data=dataFlights,col='UniqueCarrier',col_wrap=4)
plt.savefig('Retards_arribada_vs_sortida_Aerolinea.png',dpi=300,bbox_inches='tight')
```

## Més de tres variables (ArrDelay, DepDelay, AirTime i UniqueCarrier)

Es crea un dataframe nou amb el tipus de retard

```python
data_ArrDelay = dataFlights.loc[:,('UniqueCarrier','ArrDelay')]
data_ArrDelay.rename(columns={'ArrDelay': 'Time'},inplace=True)
data_ArrDelay['TimeType']='ArrDelay'
```

```python
data_DepDelay = dataFlights.loc[:,('UniqueCarrier','DepDelay')]
data_DepDelay.rename(columns={'DepDelay': 'Time'},inplace=True)
data_DepDelay['TimeType']='DepDelay'
```

```python
data_AirTime = dataFlights.loc[:,('UniqueCarrier','AirTime')]
data_AirTime.rename(columns={'AirTime': 'Time'},inplace=True)
data_AirTime['TimeType']='AirTime'
```

```python
data_Time1 = data_ArrDelay.append(data_DepDelay)
```

```python
data_Time = data_Time1.append(data_AirTime)
data_Time
```

```python
plt.figure(figsize=(21,13))
sns.set_style("darkgrid")
plt.title('Temps en cada aerolínea', size=21)
plt.xlabel('Temps',size=13)
plt.ylabel('Codí d´aerolínea',size=13)

dataTime_group = data_Time.groupby(['UniqueCarrier','TimeType'], as_index=False).sum()

sns.barplot(data=dataTime_group,y='UniqueCarrier', x='Time', hue='TimeType')

plt.savefig('Temps-Aerolineas.png',dpi=300,bbox_inches='tight')
```
